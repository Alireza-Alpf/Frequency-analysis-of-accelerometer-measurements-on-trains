# Frequency analysis of accelerometer measurements on trains

This data set contains both normal and failure datasets of metro bogies at a sampling frequency of 1024Hz, and carbody acceleration at sampling frequency of 20Hz. Dataset is collected for X, Y and Z axis vibration acceleration data of bogie and carbody. The data set includes about 50.8 minutes of normal data, 101.4 minutes of failure data, and 20 minutes of carbody data.
Two energy harvesting sensor nodes was used to collect the vibration acceleration data from two metro bogies. One bogie is in normal state and the other is in a failure state. A 3-axis accelerometer is used in each sensor node to collect 3-axis vibration acceleration data. To collect data for carbody a smartphone is used from Garbatella metro station to Jonio metro station (App: Accelerometer).
The unit of our dataset is m/s^2 and sensitivity is 100mg.

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import seaborn as sns
```

```python
#Importing csv files
acc_x_n = pd.read_csv("xn.csv")
acc_y_n = pd.read_csv("yn.csv") 
acc_z_n = pd.read_csv("zn.csv") 
acc_x_f = pd.read_csv("xf.csv")
acc_y_f = pd.read_csv("yf.csv")
acc_z_f = pd.read_csv("zf.csv")
acc_cb = pd.read_csv('carbody.csv')
acc_cb['z'] = acc_cb['z'] + 1  #accelerometer shows -1 for acceleration in z axis as default
acc_cb['x'] = acc_cb['x'] * (-1)
#Converting unit of acceleration from g to m/s^2
acc_x_n["x_data/g"] = acc_x_n["x_data/g"] * 9.80665
acc_y_n["y_data/g"] = acc_y_n["y_data/g"] * 9.80665
acc_z_n["z_data/g"] = acc_z_n["z_data/g"] * 9.80665
acc_x_f["x_data/g"] = acc_x_f["x_data/g"] * 9.80665
acc_y_f["y_data/g"] = acc_y_f["y_data/g"] * 9.80665
acc_z_f["z_data/g"] = acc_z_f["z_data/g"] * 9.80665
acc_cb['x'] = acc_cb['x'] * 9.80665
acc_cb['y'] = acc_cb['y'] * 9.80665
acc_cb['z'] = acc_cb['z'] * 9.80665
#This data had some error in collecting phase, So I had to calibrate it
acc_cb.iloc[12183:22656,1] = acc_cb.iloc[12183:22656,1] - 0.265
acc_cb.iloc[0:12183,2] = acc_cb.iloc[0:12183,2] + 1.465
acc_cb.iloc[12183:22656,2] = acc_cb.iloc[12183:22656,2] + 0.788
acc_cb.iloc[0:12183,3] = acc_cb.iloc[0:12183,3] - 0.173
acc_cb.iloc[12183:22656,3] = acc_cb.iloc[12183:22656,3] - 0.091
```

```python
#calculating max of acceleration in x-axis
print('The index max of acc_x_n',acc_x_n["x_data/g"].argmax())
print('The index max of acc_x_f',acc_x_f["x_data/g"].argmax())
print('The index max of acc_cb',acc_cb["y"].argmax())
```

```python
fig, ax = plt.subplots(3, 1,figsize=(20,20))
fig.suptitle('Acceleration of the normal bogie', fontsize=30) 
ax[0].plot(acc_x_n.index,acc_x_n["x_data/g"],color='b')
ax[1].plot(acc_y_n.index,acc_y_n["y_data/g"],color='b')
ax[2].plot(acc_z_n.index,acc_z_n["z_data/g"],color='b')
ax[0].set_xlabel("Index", size = 15)
ax[1].set_xlabel("Index", size = 15)
ax[2].set_xlabel("Index", size = 15)
ax[0].set_ylabel("The x-axis acceleration data", size = 15)
ax[1].set_ylabel("The y-axis acceleration data", size = 15)
ax[2].set_ylabel("The z-axis acceleration data", size = 15)
plt.show()
print('Maximum of acc x-axis of bogie in normal state', acc_x_n["x_data/g"].max())
print('Minimum of acc x-axis of bogie in normal state', acc_x_n["x_data/g"].min())
print('Maximum of acc y-axis of bogie in normal state', acc_y_n["y_data/g"].max())
print('Minimum of acc y-axis of bogie in normal state', acc_y_n["y_data/g"].min())
print('Maximum of acc z-axis of bogie in normal state', acc_z_n["z_data/g"].max())
print('Minimum of acc z-axis of bogie in normal state', acc_z_n["z_data/g"].min())
```

# Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

```python
fig, ax = plt.subplots(figsize=(20,15))
fig.suptitle('Acceleration of bogie and carbody(X-axis)', fontsize=25) 
ax.boxplot([acc_x_n["x_data/g"],acc_x_f["x_data/g"],acc_cb['x']])
ax.set_xticklabels(["Acc. X-axis normal", "Acc. X-axis failure","Acc. X-axis carbody"] , size = 15 )
ax.set_ylabel("Acceleration", size = 15)
plt.show()
median=np.median(acc_x_n["x_data/g"])
up_q_n = np.percentile(acc_x_n["x_data/g"],25)
print('(normal)25 percentile is', up_q_n)
up_q_n = np.percentile(acc_x_n["x_data/g"],75)
print('(normal)75 percentile is', up_q_n)
median=np.median(acc_x_f["x_data/g"])
up_q_f = np.percentile(acc_x_f["x_data/g"],25)
print('(failure)25 percentile is', up_q_f)
up_q_f = np.percentile(acc_x_f["x_data/g"],75)
print('(failure)75 percentile is', up_q_f)
median=np.median(acc_cb["x"])
up_q_f = np.percentile(acc_cb["x"],25)
print('(carbody)25 percentile is', up_q_f)
up_q_f = np.percentile(acc_cb["x"],75)
print('(carbody)75 percentile is', up_q_f)
```

# Fast Fourier Transform

A fast Fourier transform (FFT) is an algorithm that computes the discrete Fourier transform (DFT) of a sequence, or its inverse (IDFT). Fourier analysis converts a signal from its original domain (often time or space) to a representation in the frequency domain and vice versa. The DFT is obtained by decomposing a sequence of values into components of different frequencies.

```python
from scipy import pi
from scipy.fftpack import fft
```

```python
N = 22656
frequency = np.linspace (0.0, 10, int (N/2))
acc_cb_l = acc_cb['x'].tolist()
freq_data = fft(acc_cb_l)
y = 2/N * np.abs (freq_data [0:np.int (N/2)])
fig, ax = plt.subplots(figsize=(15,10))
plt.plot(frequency, y)
plt.title('Frequency domain Signal x-axis')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.xlim((0.1,10))
plt.show()
```
